Defining the keywords for my search (this can be edited and optimized laters)
building a complex query operation:

```python
keywords = [
    '("entry-level data science" OR "junior data scientist" OR "data science intern") AND ("skills" OR "requirements" OR "career advice")',
    '("starting a career in data science" OR "first data science job") AND ("skills needed" OR "how to get hired")',
    '("beginner AI skills" OR "starting AI career") AND ("hiring trends" OR "job requirements")',
    '("data science bootcamp" OR "self-taught data scientist") AND ("entry-level jobs" OR "career launch")',
    '("how to become a data scientist") AND ("no experience" OR "entry-level")',
    '("data science curriculum" OR "AI course design") AND (university OR bootcamp OR college)',
    '("machine learning syllabus" OR "AI course update") AND ("higher education" OR "academic program")',
    '("data science education" OR "AI in education") AND ("curriculum development" OR "learning outcomes")',
    '("new data science program" OR "ML certificate course") AND (Coursera OR university OR bootcamp)',
    '("teaching AI in schools" OR "data science in universities") AND ("course content" OR "syllabus update")'
]
```

install packages and scrapping articles:

```python
import pandas
import tqdm
import gnews
import time
import pandas as pd
from newspaper import Article
from googlesearch import search
```

Googlesearch import search generates accessable urls, and scrappes articles effectively

```python
# -------------------------------
# 2. Google Search + Article Scrape
# -------------------------------

num_results = 10
articles_data = []

for keyword in keywords:
    print(f"\nüîç Searching for: {keyword}")
    urls = search(keyword, num_results=num_results, lang="en")
    
    for url in urls:
        print(f"‚úÖ Found: {url}")
        try:
            article = Article(url)
            article.download()
            article.parse()

            if len(article.text.strip()) < 200:
                print("‚ö†Ô∏è Article too short, skipping.")
                continue

            articles_data.append({
                "keyword": keyword,
                "title": article.title,
                "url": url,
                "text": article.text
            })

            time.sleep(1.5)  # polite delay between scrapes

        except Exception as e:
            print(f"‚ùå Failed to process {url}: {e}")
            continue

# -------------------------------
# 3. Save to CSV
# -------------------------------

df = pd.DataFrame(articles_data)
df.to_csv("tech_trends_news.csv", index=False)
print("\nüìÅ Results saved to 'tech_trends_news.csv'")
```
